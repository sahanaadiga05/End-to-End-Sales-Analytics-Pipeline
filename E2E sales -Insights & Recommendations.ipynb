# End-to-End Sales Analytics Pipeline — Insights & Recommendations

## Key Business Insights

### 1. Technology drives highest profit
- Technology revenue = $836,154
- Technology profit margin = 17%
- Technology contributes 45% of total profit
- Highest margin sub-category = Copiers (37%)
- Lowest margin sub-category = Machines (3%)

### 2. Furniture is a losing category
- Furniture has high sales but only 3% margin
- Tables profit margin = NEGATIVE (-8%)
- Bookcases profit margin = NEGATIVE (-3%)
- Heavy discounting on Furniture = main cause
- Every 10% discount on Furniture = 5% margin loss

### 3. Discounts are destroying profit
- No discount orders = 31% profit margin
- Low discount (1-20%) = 22% profit margin
- Medium discount (21-40%) = 5% profit margin
- High discount (41%+) = NEGATIVE profit margin
- 20% of all orders have high discounts
- These orders lose money on every sale

### 4. West region is most profitable
- West region = highest revenue and profit
- West profit margin = 20%+
- South region = lowest profit margin
- South has high discounting = main cause
- Central region = moderate performance

### 5. Consumer segment dominates revenue
- Consumer = 51% of total revenue
- Corporate = 30% of total revenue
- Home Office = 19% of total revenue
- But Corporate has BETTER profit margin
- Corporate avg order value = highest

### 6. Office Supplies has best volume
- Office Supplies = highest order count
- But lower avg order value than Technology
- Good repeat purchase category
- Labels and Paper = highest volume items

### 7. Sub-category performance gap
- Copiers = highest profit margin (37%)
- Phones = highest revenue ($330,007)
- Tables = biggest loss maker (-$17,725)
- Bookcases = second biggest loss (-$3,472)
- Machines = high sales but very low margin

### 8. High value orders drive revenue
- Orders above $500 = 30% of total revenue
- Corporate segment = most high value orders
- Technology category = most $500+ orders
- West region = most high value orders

## Business Recommendations (With Numbers)

### 1. Cap all discounts at 20% maximum
- Current: 20% orders have 40%+ discount
- These orders have NEGATIVE profit margin
- Capping discounts at 20% could increase
  overall profit margin by 8-10%
- Expected additional profit = $25,000-30,000/year
- Start with Furniture category first

### 2. Discontinue or reprice Tables
- Tables lose $17,725 every year
- Either increase Tables price by 20%
- Or discontinue low margin Table variants
- Could save $17,000+ in annual losses
- Redirect shelf space to Copiers and Phones

### 3. Focus marketing on Technology
- Technology = 45% of total profit
- Increasing Technology marketing by 10%
  could grow revenue by $50,000+
- Specifically push Copiers (37% margin)
- Bundle Phones with accessories

### 4. Target Corporate segment more
- Corporate has better profit margin
  than Consumer segment
- Corporate avg order value is highest
- Increasing Corporate sales by 10%
  could improve overall margin by 2-3%
- Assign dedicated sales team for B2B
- Expected revenue increase = $40,000+

### 5. Fix South region performance
- South has 3rd highest revenue
  but lowest profit margin
- High discounting in South = main cause
- Reduce South region discounts by 15%
- Could add $15,000 in annual profit
- Investigate competitor pricing in South

### 6. Invest more in West region
- West has highest profit margin
- Best ROI on marketing spend
- Increasing West marketing by 15%
  could generate $40,000 additional revenue
- Focus on Technology in West region

### 7. Improve Office Supplies margins
- Office Supplies has high volume
  but moderate margins
- Bundle Office Supplies with Technology
- Create corporate bulk order discounts
- Could increase avg order value by 12%
